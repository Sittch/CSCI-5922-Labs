```python
## Scikit-Learn Python Library (https://scikit-learn.org/stable/)
## Outline:
#Part 1. Load datasets
#Part 2. Create the train/test split and preprocess
#Part 3. Train NN
#Part 4. Evaluation metrics + Confusion Matrix
#Part 5. Plot accuracy vs epoch
```

```python
# Part 1: Load datasets
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.data
y = digits.target
```

```python
#2. Part 2: Create the train/test split and preprocess
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)

# Standardize data
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
stdsc.fit(X_train)
X_train_std = stdsc.transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
# Train Neural Networks
from sklearn.neural_network import MLPClassifier

n_hidden_nodes = 10
mlp1 = MLPClassifier(activation = 'identity', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes])
mlp1.fit(X_train_std, y_train)
mlp2 = MLPClassifier(activation = 'identity', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes, n_hidden_nodes])
mlp2.fit(X_train_std, y_train)
mlp3 = MLPClassifier(activation = 'identity', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes, n_hidden_nodes, n_hidden_nodes])
mlp3.fit(X_train_std, y_train)
mlp4 = MLPClassifier(activation = 'logistic', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes])
mlp4.fit(X_train_std, y_train)
mlp5 = MLPClassifier(activation = 'logistic', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes, n_hidden_nodes])
mlp5.fit(X_train_std, y_train)
mlp6 = MLPClassifier(activation = 'logistic', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes, n_hidden_nodes, n_hidden_nodes])
mlp6.fit(X_train_std, y_train)
mlp7 = MLPClassifier(activation = 'tanh', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes])
mlp7.fit(X_train_std, y_train)
mlp8 = MLPClassifier(activation = 'tanh', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes, n_hidden_nodes])
mlp8.fit(X_train_std, y_train)
mlp9 = MLPClassifier(activation = 'tanh', hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes, n_hidden_nodes, n_hidden_nodes])
mlp9.fit(X_train_std, y_train)

print("mlp1:")
print("Activation Function: {}".format(mlp1.activation))
print("Accuracy score on the test set: {}".format(mlp1.score(X_test_std, y_test)))
print("mlp2")
print("Activation Function: {}".format(mlp2.activation))
print("Accuracy score on the test set: {}".format(mlp2.score(X_test_std, y_test)))
print("mlp3")
print("Activation Function: {}".format(mlp3.activation))
print("Accuracy score on the test set: {}".format(mlp3.score(X_test_std, y_test)))
print("mlp4")
print("Activation Function: {}".format(mlp4.activation))
print("Accuracy score on the test set: {}".format(mlp4.score(X_test_std, y_test)))
print("mlp5")
print("Activation Function: {}".format(mlp5.activation))
print("Accuracy score on the test set: {}".format(mlp5.score(X_test_std, y_test)))
print("mlp6")
print("Activation Function: {}".format(mlp6.activation))
print("Accuracy score on the test set: {}".format(mlp6.score(X_test_std, y_test)))
print("mlp7")
print("Activation Function: {}".format(mlp7.activation))
print("Accuracy score on the test set: {}".format(mlp7.score(X_test_std, y_test)))
print("mlp8")
print("Activation Function: {}".format(mlp8.activation))
print("Accuracy score on the test set: {}".format(mlp8.score(X_test_std, y_test)))
print("mlp9")
print("Activation Function: {}".format(mlp9.activation))
print("Accuracy score on the test set: {}".format(mlp9.score(X_test_std, y_test)))
```

```python
#Part 4: Evaluation Metrics and Confusion Matrix

#1: Eval metrics

from sklearn import metrics

print("mlp1")
y_predicted_mlp1 = mlp1.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp1))
print("mlp2")
y_predicted_mlp2 = mlp2.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp2))
print("mlp3")
y_predicted_mlp3 = mlp3.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp3))
print("mlp4")
y_predicted_mlp4 = mlp4.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp4))
print("mlp5")
y_predicted_mlp5 = mlp5.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp5))
print("mlp6")
y_predicted_mlp6 = mlp6.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp6))
print("mlp7")
y_predicted_mlp7 = mlp7.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp7))
print("mlp8")
y_predicted_mlp8 = mlp8.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp8))
print("mlp9")
y_predicted_mlp9 = mlp9.predict(X_test_std)
print(metrics.classification_report(y_test, y_predicted_mlp9))
```

```python
#2: Confusion Matrix

import seaborn as sns
import matplotlib.pyplot as plt

print("mlp1")
mat1 = metrics.confusion_matrix(y_test, y_predicted_mlp1)
sns.heatmap(mat1.T, square = True, annot = True, fmt = 'd', cbar = False)
plt.xlabel("True label")
plt.ylabel("Predicted label")
```
